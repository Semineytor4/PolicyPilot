# Customer Churn Prediction Using Classical Machine Learning on Tabular Customer Data

## 1. Einleitung und Motivation
Kundenabwanderung (Customer Churn) ist für viele Unternehmen wirtschaftlich kritisch, da die Akquise neuer Kunden in der Regel deutlich teurer ist als die Bindung bestehender Kunden. Datengetriebene Methoden ermöglichen es, kündigungsgefährdete Kunden frühzeitig zu identifizieren, sodass gezielte Retentionsmaßnahmen (z. B. Angebote, Service-Interventionen) geplant werden können.

Ziel dieser Arbeit ist die Entwicklung, der Vergleich und die Evaluation klassischer Machine-Learning-Modelle zur Vorhersage von Kundenabwanderung auf Basis tabellarischer Kundendaten. Neben der Vorhersagequalität werden Interpretierbarkeit und Rechenaufwand betrachtet, um die praktische Einsetzbarkeit in Unternehmenskontexten zu bewerten. Die Ergebnisse werden so aufbereitet, dass sie für Fachanwender und Nicht-Experten verständlich sind.

## 2. Forschungsfrage und Zielsetzung

### 2.1 Zentrale Forschungsfrage
Wie gut lässt sich die Abwanderung von Kunden (Customer Churn) mithilfe klassischer Machine-Learning-Verfahren auf Basis tabellarischer Kundendaten vorhersagen, und welcher Algorithmus bietet dabei die beste Kombination aus Vorhersagequalität, Interpretierbarkeit und Rechenaufwand?

### 2.2 Thematische Forschungsfragen
- Welche Merkmale (z. B. Vertragsdauer, Kosten, Service-Nutzung) beeinflussen die Wahrscheinlichkeit einer Kundenkündigung besonders stark?
- Wie unterscheiden sich klassische Machine-Learning-Algorithmen in ihrer Vorhersageleistung bei tabellarischen Daten?
- Welche Rolle spielt die Datenvorverarbeitung (Encoding, Skalierung, Missing Values) für die Modellleistung?
- Wie lassen sich Ergebnisse für Fachanwender verständlich interpretieren und visualisieren?
- Welche Limitationen und Unsicherheiten bestehen bei der Vorhersage von Kundenabwanderung?

### 2.3 (Optional) Arbeitshypothesen
- H1: Vertragsmerkmale (z. B. Vertragslaufzeit/Contract Type, Tenure) haben einen stärkeren Einfluss auf Churn als demografische Merkmale.
- H2: Ensemble-Modelle (Random Forest, Gradient Boosting) erzielen höhere Vorhersageleistung als lineare Modelle, sind jedoch schwerer interpretierbar.
- H3: Saubere Vorverarbeitung (korrektes Encoding, Skalierung für lineare Modelle, Leakage-freies Vorgehen) verbessert die Generalisierung messbar.

Die Zielgruppe sind Unternehmen, Datenanalysten, Studierende sowie Fachabteilungen (z. B. Marketing, Customer Success), die Kundenbindung datenbasiert unterstützen möchten.

### 3.1 Anwendungsnutzen
- Frühzeitige Identifikation kündigungsgefährdeter Kunden

- Verständliche Visualisierung und Interpretation der wichtigsten Einflussfaktoren

## 4. Datengrundlage

### 4.1 Datensatz
Als Datengrundlage dient ein öffentlich verfügbarer, reproduzierbarer Datensatz, z. B. der **Telco Customer Churn** Datensatz (IBM Sample Data). Er enthält anonymisierte Informationen zu Kunden, Vertragsdaten, genutzten Services und Kündigungsstatus (Churn: Yes/No).

### 4.2 Eigenschaften und Transparenz
- Tabellarische CSV-Daten (strukturierte Daten)
- Anonymisierte Daten, keine personenbezogenen Echt-Daten
- Häufig in Lehr- und Forschungsprojekten genutzt, gut vergleichbar
- Frei verfügbar und damit reproduzierbar (Replikation durch Dritte möglich)

### 4.3 Datenverständnis und mögliche Herausforderungen
- Kategoriale Features (z. B. Vertragstyp, Payment Method, Services)
- Numerische Features (z. B. Tenure, MonthlyCharges, TotalCharges)
- Potenzielle Datenqualitätsprobleme (z. B. fehlende Werte, Datentypen wie TotalCharges als String)
- Klassenungleichgewicht möglich (Churn i. d. R. seltener als No-Churn)

## 5. Methodik und Machine-Learning-Ansatz

### 5.1 Problemformulierung
Das Problem wird als **überwachtes Lernproblem** (Supervised Learning) formuliert: binäre Klassifikation in:
- **Churn** (kündigt)
- **No Churn** (kündigt nicht)

### 5.2 Pipeline-Überblick (Leakage-frei)
Der ML-Prozess wird reproduzierbar als Pipeline umgesetzt:
1. Explorative Datenanalyse (EDA)
2. Train/Validation/Test-Strategie (mit Stratifizierung)
3. Vorverarbeitung (Encoding, Skalierung, Missing Values)
4. Modelltraining und Hyperparameter-Tuning
5. Evaluation mit geeigneten Metriken
6. Modellvergleich und Interpretierbarkeit
7. Aufbereitung für WebApp (Inference-Pipeline)

Wichtig: Vorverarbeitung erfolgt **innerhalb einer scikit-learn Pipeline** (z. B. `ColumnTransformer`), um Datenleckage zu vermeiden.

### 5.3 Datenvorverarbeitung
- **Kategoriale Variablen:** One-Hot-Encoding (oder Ordinal-Encoding, falls sinnvoll)
- **Numerische Variablen:** Skalierung (z. B. StandardScaler) für lineare Modelle (LogReg, SVM)
- **Missing Values:** Imputation (z. B. Median für numerisch, häufigste Kategorie für kategorial)
- **Outlier/Skew:** optional prüfen; Transformation nur wenn begründet

### 5.4 Umgang mit Klassenungleichgewicht
Da Churn oft seltener ist, werden Maßnahmen berücksichtigt:
- Stratifizierte Splits und Cross-Validation
- Metriken, die Imbalance berücksichtigen (F1, Recall, PR-AUC)
- Optional: `class_weight='balanced'` oder Threshold-Tuning
- Optional: Resampling (z. B. SMOTE) **nur innerhalb des CV-Prozesses** (Leakage vermeiden)

## 6. Modelle und Vergleich

### 6.1 Baseline
Als Referenz wird ein **DummyClassifier** (Mehrheitsklasse) verwendet. Dadurch wird sichtbar, ob ML-Modelle einen echten Mehrwert gegenüber trivialen Strategien liefern.

### 6.2 Klassische Modelle (Kandidaten)
- **Logistische Regression** (interpretierbar, schnell, starke Baseline)
- **Support Vector Machine (linear/RBF)** (optional, oft gut bei sauberer Skalierung)
- **Random Forest** (robust, nichtlinear, Feature Importance)
- **Gradient Boosting** (z. B. GradientBoostingClassifier, XGBoost/LightGBM optional)  
  Hinweis: Fokus liegt auf klassischen, weit verbreiteten Verfahren für tabellarische Daten.

### 6.3 Rechenaufwand
Rechenaufwand wird messbar gemacht durch:
- Trainingszeit (z. B. gemittelt über CV-Folds)
- Inferenzzeit pro Beispiel (für WebApp relevant)
- Modellkomplexität (z. B. Anzahl Parameter/Bäume, Tiefe)

## 7. Evaluationsdesign

### 7.1 Validierungsstrategie
- Primär: **Stratified K-Fold Cross-Validation** für robustere Leistungsabschätzung
- Optional: **Nested Cross-Validation** für fairen Vergleich bei Hyperparameter-Tuning
- Finales Modell: Training auf Train+Val, abschließender Test auf Holdout-Testset

### 7.2 Metriken (Performance und Business-Relevanz)
Da Accuracy bei Imbalance irreführend sein kann, werden mehrere Metriken genutzt:
- **ROC-AUC** (Trennschärfe, schwellenunabhängig)
- **F1-Score** (Balance zwischen Precision und Recall, wichtig bei Churn-Klasse)
- **Precision & Recall** (z. B. Recall wichtig, um möglichst viele Churner zu finden)
- **PR-AUC** (bei starker Imbalance oft aussagekräftiger als ROC-AUC)
- **Confusion Matrix** (anschauliche Fehleranalyse)
- Optional: **Kalibrierung** (Calibration Curve, Brier Score) für verlässliche Wahrscheinlichkeiten

### 7.3 Modellselektion
Das „beste“ Modell wird nicht nur nach einer Kennzahl gewählt, sondern nach:
- Vorhersagequalität (primäre Metrik, z. B. PR-AUC oder F1)
- Interpretierbarkeit (z. B. LogReg-Koeffizienten vs. Permutation Importance)
- Rechenaufwand (Trainings-/Inferenzzeit)

## 8. Interpretierbarkeit und Visualisierung

## 11. Limitationen und Risiken

### 11.1 Validität und Generalisierung
- Datensatz repräsentiert möglicherweise nicht alle Unternehmenskontexte
- Ergebnisse sind korrelativ, nicht kausal (keine Garantie, dass Maßnahmen wirken)

### 11.2 Daten- und Modellrisiken
- Klassenimbalance und Schwellenwahl beeinflussen Ergebnisse
- Bias-Risiken (z. B. demografische Merkmale) und Fairness-Aspekte
- Konzeptdrift: Kundenverhalten kann sich über Zeit ändern (Modell muss ggf. nachtrainiert werden)

### 11.3 Unsicherheit
- Wahrscheinlichkeiten sollten (optional) kalibriert werden, um interpretierbar zu bleiben
- Fehlklassifikationen haben Kosten: Business-Kostenmatrix sollte perspektivisch berücksichtigt werden

## 12. Erwartete Ergebnisse
Erwartet wird, dass Ensemble-Modelle (Random Forest/Gradient Boosting) eine höhere Vorhersageleistung erzielen, während lineare Modelle (LogReg) eine bessere Interpretierbarkeit und geringeren Rechenaufwand bieten. Die finale Auswahl erfolgt anhand der definierten Performance-Metriken, Interpretierbarkeit und Effizienz.

## 13. Zuordnung zum QUA³CK-Prozessmodell

Zur strukturierten Entwicklung des Machine-Learning-Modells wird das
QUA³CK-Prozessmodell angewendet. Die einzelnen Phasen werden im Projekt
wie folgt umgesetzt:

| Phase (QUA³CK) | Umsetzung im Projekt |
|---------------|----------------------|
| Q – Question | Definition der Forschungsfrage und Zielsetzung |
| U – Understanding | Explorative Datenanalyse (EDA) mit pandas |
| A – Algorithm Selection | Auswahl geeigneter Klassifikationsmodelle (LogReg, Random Forest, Gradient Boosting) |
| A – Data Adaption | Datenvorverarbeitung (Encoding, Skalierung, Imputation) |
| A – Parameter Adjustment | Hyperparameter-Tuning mittels Cross-Validation |
| C – Conclusion & Comparison | Vergleich der Modelle anhand ROC-AUC, F1-Score und Confusion Matrix |
| K – Knowledge Transfer | Visualisierung der Ergebnisse und Deployment mittels Streamlit-WebApp |
