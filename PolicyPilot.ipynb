 #  📌 Policypilot – KI-gestützter Versicherungs-Chatbot

 ## Policypilot – Ein KI-gestützter Chatbot zur Beantwortung von Versicherungs Fragen

## Zentrale Forschungsfrage

Wie kann ein KI-gestützter Chatbot Versicherungs- und API-bezogene Fragestellungen zuverlässig, kontextbezogen und effizient beantworten, indem interne API-Schnittstellen und bereitgestellte Dokumentationen intelligent eingebunden werden?

## Thematische Forschungsfragen

Wie lassen sich interne Versicherungs-APIs sicher und kontrolliert in einen KI-Chatbot integrieren?

Wie kann Dokumentation (z. B. API-Specs, technische Beschreibungen) für ein Sprachmodell nutzbar gemacht werden?

Welche Vorteile bietet ein dialogbasierter Zugang zu Versicherungswissen gegenüber klassischer Dokumentation?

Welche Risiken bestehen (Halluzinationen, falsche Antworten) und wie können diese minimiert werden?

## 🎯 Zielgruppe und Anwendungsnutzen

Die primäre Zielgruppe von Policypilot sind API-Konsumenten, interne Entwickler, Fachabteilungen sowie Mitarbeitende im Versicherungs- und API-Umfeld, die regelmäßig Fragen zu Versicherungsprodukten, Schnittstellen oder technischer Dokumentation haben.

Der Chatbot dient als:

Erstkontakt-Anlaufstelle für Versicherungs- und API-Fragen

Zeitersparnis für das API-Team durch automatisierte Beantwortung wiederkehrender Fragen

Interaktive Alternative zur klassischen API-Dokumentation

Wissensassistent für komplexe Zusammenhänge im Versicherungsumfeld

## 🗂️ Datengrundlage – transparent und praxisnah

Die Datengrundlage von Policypilot besteht aus:

internen API-Dokumentationen (z. B. OpenAPI/Swagger, Markdown, PDFs)

fachlichen Versicherungsinformationen

strukturierten API-Antworten aus internen Services

Die KI greift nicht frei auf das Internet zu, sondern nutzt gezielt:

bereitgestellte Dokumente (Retrieval)

Live-API-Abfragen zur Beantwortung konkreter technischer Fragen

Dies erhöht die Verlässlichkeit, Nachvollziehbarkeit und Datensicherheit.

## 🤖 Machine-Learning-Prozess und Modelllogik

Policypilot basiert nicht auf einem selbst trainierten Modell, sondern auf einem Large Language Model (LLM), das über folgende Mechanismen gesteuert wird:

Prompt Engineering zur klaren Rollen- und Aufgabenbeschreibung

### Retrieval-Augmented Generation (RAG):

relevante Dokumente werden vor der Antwort aus einem Wissensspeicher geladen

API-Tool-Calling:

bei technischen Fragen ruft die KI gezielt interne APIs auf

Antwortvalidierung:

Einschränkung auf bekannte Datenquellen zur Vermeidung von Halluzinationen

Diese Architektur ist ressourcenschonend und praxisnah.
