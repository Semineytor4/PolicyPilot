# 1. Datenbeschreibung

Der Datensatz besteht aus folgenden Variablen:

- `customerID`: eindeutige Kunden-ID  
- `gender`: Geschlecht des Kunden  
- `SeniorCitizen`: Kennzeichnung, ob Kunde Senior ist (0 = nein, 1 = ja)  
- `Partner`: Partnerstatus  
- `Dependents`: Ob der Kunde abhängige Personen hat  
- `tenure`: Vertragsdauer in Monaten  
- `PhoneService / MultipleLines`: Telefonnutzung  
- `InternetService`: Internetanbieter (DSL, Fiber, None)  
- `OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport`: Zusatzservices  
- `StreamingTV / StreamingMovies`: Streamingdienste  
- `Contract`: Vertragsart (monatlich, 1 Jahr, 2 Jahre)  
- `PaperlessBilling`: Online-Rechnung  
- `PaymentMethod`: Zahlungsmethode  
- `MonthlyCharges`: Monatliche Kosten  
- `TotalCharges`: Gesamtkosten  
- `Churn`: Zielvariable (Yes = Kunde kündigt, No = Kunde bleibt)  

---

# 2. Explorative Datenanalyse (EDA)

## 2.1 Verteilung der Zielvariable (Churn)

Die Analyse zeigt, dass der Großteil der Kunden im Datensatz bleibt, während ein signifikanter Anteil kündigt.  
Diese **unausgeglichene Klassenverteilung** ist typisch für Churn-Datensätze und muss bei späteren Modellen berücksichtigt werden.

---

## 2.2 Numerische Merkmale

- **Vertragsdauer (tenure)**  
  Kunden mit kurzer Vertragsdauer kündigen häufiger. Langjährige Kunden sind deutlich loyaler.

- **Monatliche Kosten (MonthlyCharges)**  
  Höhere monatliche Kosten sind mit einer höheren Kündigungswahrscheinlichkeit verbunden.  
  Dies deutet darauf hin, dass Preis ein wichtiger Faktor für Churn ist.

---

## 2.3 Kategorische Merkmale

- **Vertragsart (Contract)**  
  Kunden mit monatlichen Verträgen kündigen deutlich häufiger als Kunden mit 1- oder 2-Jahresverträgen.  
  Langfristige Verträge erhöhen die Kundenbindung.

- **InternetService**  
  Fiber-Internet-Kunden zeigen eine höhere Kündigungsrate im Vergleich zu DSL-Kunden.

- **Zusatzservices (OnlineSecurity, TechSupport)**  
  Kunden ohne Sicherheits- oder Supportdienste kündigen häufiger. Zusatzservices erhöhen offenbar die Kundenzufriedenheit.

---

# 3. Korrelationsanalyse

Die Korrelationsanalyse zeigt:

- `tenure` ist stark negativ mit Churn korreliert (lange Vertragsdauer → weniger Kündigung)  
- `MonthlyCharges` ist positiv mit Churn korreliert (höhere Kosten → mehr Kündigung)  
- `TotalCharges` ist indirekt mit `tenure` verbunden  

Dies bestätigt die Ergebnisse der explorativen Analyse.

---

# 5. Fazit (wissenschaftlich)

Die Analyse zeigt, dass **Vertragsdauer, Vertragsart und monatliche Kosten** zentrale Einflussfaktoren für Kundenabwanderung sind.  
Besonders Kunden mit kurzen Vertragslaufzeiten und hohen monatlichen Gebühren haben eine erhöhte Kündigungswahrscheinlichkeit.  

Zusatzservices wie **Online-Security** und **Tech-Support** wirken churn-reduzierend.  
Diese Erkenntnisse können Unternehmen helfen, **gezielte Maßnahmen zur Kundenbindung** zu entwickeln.
