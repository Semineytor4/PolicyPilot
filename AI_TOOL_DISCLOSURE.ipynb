```python
Erklärung zur Nutzung von KI-Werkzeugen im Machine-Learning-Projekt

Dieses Dokument beschreibt den Einsatz von KI-basierten Werkzeugen im Rahmen des Machine-Learning-Projekts. Die Gliederung folgt dem QUA³CK-Prozessmodell, um transparent darzustellen, in welcher Phase des Projekts welche Werkzeuge zu welchem Zweck verwendet wurden.

Das QUA³CK-Modell ist ein iterativer Prozess für die Entwicklung von Machine-Learning-Lösungen und steht für:

## Q – Question (Problemdefinition)

U – Understanding (Datenverständnis)

A – Algorithm Selection (Algorithmusauswahl)

A – Data Adaption (Datenvorverarbeitung)

A – Parameter Adjustment (Hyperparameter-Optimierung)

C – Conclusion & Comparison (Modellvergleich und Interpretation)

K – Knowledge Transfer (Visualisierung und Dokumentation)

Alle durch KI-Systeme generierten Ergebnisse (z. B. Code, Analysen, Texte) wurden von mir als verantwortlichem Entwickler kritisch geprüft, validiert und angepasst. Die finale Verantwortung für das Projekt liegt vollständig bei mir.

    | Phase (QUA³CK)              | KI-Tool (Version)                              | Zweck                          | Beispielhafter Prompt / Anwendungsfall                                                                                |
| --------------------------- | ---------------------------------------------- | ------------------------------ | --------------------------------------------------------------------------------------------------------------------- |
| Q – Question                | GitHub Copilot (GPT-4o), Google Gemini 1.5 Pro | Recherche Datensätze, Konzept  | „Finde aktuelle öffentliche Datensätze zur Kundenabwanderung (Customer Churn). Liste Features und Zielvariablen auf.“ |
| U – Understanding           | OpenAI GPT-4o                                  | Explorative Datenanalyse (EDA) | „Erstelle Python-Code mit pandas und seaborn zur Visualisierung der Churn-Verteilung und Korrelationen.“              |
| A – Algorithm Selection     | Google Gemini 1.5 Pro                          | Algorithmenauswahl             | „Vergleiche Logistic Regression, Random Forest und Gradient Boosting für tabellarische Daten.“                        |
| A – Data Adaption           | GitHub Copilot Workspace                       | scikit-learn Preprocessing     | „Implementiere eine Pipeline mit OneHotEncoder und StandardScaler.“                                                   |
| A – Parameter Adjustment    | GitHub Copilot (GPT-4o)                        | Hyperparameter-Tuning          | „Erstelle GridSearchCV für Random Forest und logge Accuracy und F1-Score.“                                            |
| C – Conclusion & Comparison | OpenAI GPT-4o                                  | Modellvergleich                | „Vergleiche Modelle anhand ROC-AUC und Confusion Matrix.“                                                             |
| K – Knowledge Transfer      | OpenAI GPT-4o                                  | Streamlit-Dashboard            | „Generiere eine Streamlit-App zur Churn-Vorhersage einzelner Kunden.“                                                 |
| K – Knowledge Transfer      | DeepL Write, OpenAI GPT-4o                     | Dokumentation & Übersetzung    | „Formuliere eine Zusammenfassung der Ergebnisse für README.md.“     |


    | Datensatz            | Quelle                               | Beschreibung                                                                         | Verwendungszweck                                   | Link                                                                                                                                                                                       |
| -------------------- | ------------------------------------ | ------------------------------------------------------------------------------------ | -------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| Telco Customer Churn | IBM Sample Data / Kaggle             | Datensatz mit 7.043 Kunden, inkl. Demografie, Services und Kündigungsstatus (Churn). | Hauptdatensatz für Training, Validierung und EDA.  | [https://www.kaggle.com/datasets/blastchar/telco-customer-churn](https://www.kaggle.com/datasets/blastchar/telco-customer-churn)                                                           |
| Bank Customer Churn  | Kaggle (Public Domain)               | Daten von 10.000 Bankkunden mit Kredit-Score, Balance und Produktanzahl.             | Validierung der Pipeline-Generalisierbarkeit.      | [https://www.kaggle.com/datasets/shrutimechlearn/churn-modelling](https://www.kaggle.com/datasets/shrutimechlearn/churn-modelling)                                                         |
| Synthetic Churn Data | scikit-learn (synthetisch generiert) | Künstlich erzeugter Datensatz mit simulierten Kundenmerkmalen.                       | Lasttests und Benchmarking der Modellinferenzzeit. | [https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html) |
```
