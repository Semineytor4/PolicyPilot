 #  📌 Policypilot – KI-gestützter Versicherungs-Chatbot

 ## Policypilot – Ein KI-gestützter Chatbot zur Beantwortung von Versicherungs Fragen

## Zentrale Forschungsfrage

Wie kann ein KI-gestützter Chatbot Versicherungs- und API-bezogene Fragestellungen zuverlässig, kontextbezogen und effizient beantworten, indem interne API-Schnittstellen und bereitgestellte Dokumentationen intelligent eingebunden werden?

Diese Forschungsfrage zielt darauf ab, die Integration von KI in den Arbeitsalltag von API-Teams zu untersuchen. Außerdem soll geprüft werden, wie ein Chatbot Fehler minimiert und verlässliche Antworten liefert, ohne dass die Nutzer tief in die Dokumentation einsteigen müssen.

## Thematische Forschungsfragen

Integration von APIs: Wie können interne Versicherungs-APIs sicher in einen KI-Chatbot integriert werden?

Dokumentations-Nutzung: Wie kann die bereitgestellte API- und Versicherungsdokumentation effektiv für die KI nutzbar gemacht werden?

Dialogbasierte Wissensvermittlung: Welche Vorteile bietet ein interaktiver Chat gegenüber statischen Dokumentationen?

Qualität und Sicherheit der Antworten: Welche Mechanismen minimieren Halluzinationen und falsche Antworten?

Nutzerfreundlichkeit: Wie kann der Chatbot so gestaltet werden, dass er für Entwickler und Fachabteilungen intuitiv bedienbar ist?

## 🎯 Zielgruppe und Anwendungsnutzen

Die primäre Zielgruppe von Policypilot sind interne API-Konsumenten, Entwicklerteams sowie Mitarbeitende, die regelmäßig mit Versicherungsprodukten oder Schnittstellen arbeiten.

Nutzen des Chatbots:

Zeitersparnis: Automatisierte Antworten auf wiederkehrende Fragen

Wissensvermittlung: Interaktive Alternative zu klassischen Dokumentationen

Fehlerreduktion: Minimierung von Fehlinterpretationen in der API-Nutzung

Zentrale Informationsquelle: Alles aus einer Hand – API, Dokumentation und Fachwissen

## 🗂️ Datengrundlage – transparent und praxisnah

Policypilot basiert ausschließlich auf vertrauenswürdigen, internen Quellen:

API-Dokumentationen: OpenAPI/Swagger, Markdown, PDFs

Fachinformationen zu Versicherungen: interne Handbücher und Leitfäden

Strukturierte API-Daten: Echtzeit-Antworten aus internen Services

Die KI nutzt keine öffentlichen Internetquellen, wodurch die Antworten sicher, nachvollziehbar und datenschutzkonform sind. Durch Retrieval-Augmented Generation (RAG) werden nur relevante Informationen abgerufen, bevor die Antwort generiert wird.

## 🤖 Machine-Learning-Prozess und Modelllogik

Die Architektur von Policypilot basiert auf einem Large Language Model (LLM), das gezielt auf interne Datenquellen zugreift:

Prompt Engineering: Rollen- und Aufgabenbeschreibung für die KI

RAG-Verfahren: Abfrage relevanter Dokumente und Einbettung in die Antwort

API-Tool-Calling: Bei technischen Fragen ruft die KI direkt die passenden APIs auf

Antwortvalidierung: Nur Daten aus bekannten Quellen werden genutzt, um Halluzinationen zu vermeiden

Durch diese Vorgehensweise kann ressourcenschonend und zuverlässig auf komplexe Fragen geantwortet werden.

### Retrieval-Augmented Generation (RAG):

relevante Dokumente werden vor der Antwort aus einem Wissensspeicher geladen

API-Tool-Calling:

bei technischen Fragen ruft die KI gezielt interne APIs auf

Antwortvalidierung:

Einschränkung auf bekannte Datenquellen zur Vermeidung von Halluzinationen

Diese Architektur ist ressourcenschonend und praxisnah.

## 🖥️ Umsetzungskonzept und WebApp-Skizze

Technologien: Python, Jupyter Notebook, Streamlit

### Struktur der Notebooks:

01_konzept.ipynb: Konzept, Ziel, Forschungsfragen

02_architektur.ipynb: Technische Architektur, API-Schema

03_api_anbindung.ipynb: API-Calls, Authentifizierung, Beispiele

04_chatbot_logik.ipynb: Prompt-Design, RAG, Antwortlogik

05_streamlit_app.ipynb: Frontend/Chat-Interface

#### WebApp-Skizze:

User → Streamlit UI → KI-Logik
  ├─ Dokumenten-Retriever → Antwort
  └─ API-Abfrage → Antwort
→ UI zeigt Antwort inkl. Quellen

Anzeige von Antworttext und Quellen

Optional: Historie früherer Fragen
